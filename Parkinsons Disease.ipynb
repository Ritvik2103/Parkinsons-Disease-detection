# Split data into features (X) and target (y) 

```python
X = df.drop(['status', 'name'], axis=1)
y = df['status']
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Standard scaling

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Importing Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# Function For Model Evaluation

```python
def evaluate_model(model_name, y_true, y_pred):
    print(f"Model: {model_name}")
    print(f"Accuracy: {accuracy_score(y_true, y_pred)}")
    print("Classification Report:")
    print(classification_report(y_true, y_pred))
    print("Confusion Matrix:")
    cf_matrix=confusion_matrix(y_true, y_pred)
    sns.heatmap(cf_matrix, annot=True)
    plt.show()
```

# Logistic Regression

```python
logreg = LogisticRegression(max_iter=5000)
logreg.fit(X_train_scaled, y_train)
y_pred_logreg = logreg.predict(X_test_scaled)
```

```python
evaluate_model("Logistic Regression", y_test, y_pred_logreg)
```

### Finding Optimal Parameters 

```python
param_grid_logreg = {'C': [0.01, 0.1, 1, 10, 100]}
grid_logreg = GridSearchCV(logreg, param_grid_logreg, cv=5)
grid_logreg.fit(X_train_scaled, y_train)
y_pred_logreg = grid_logreg.predict(X_test_scaled)
```

```python
grid_logreg.best_estimator_
```

```python
evaluate_model("Logistic Regression", y_test, y_pred_logreg)
```

# K-Nearest Neighbors (KNN)

```python
knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)
```

```python
evaluate_model("K-Nearest Neighbors (KNN)", y_test, y_pred_knn)
```

### Finding Optimal Parameters 

```python
param_grid_knn = {'n_neighbors': [3, 5, 7, 9, 10, 20]}
grid_knn = GridSearchCV(knn, param_grid_knn, cv=5)
grid_knn.fit(X_train_scaled, y_train)
y_pred_knn = grid_knn.predict(X_test_scaled)
```

```python
grid_knn.best_estimator_
```

```python
evaluate_model("K-Nearest Neighbors (KNN)", y_test, y_pred_knn)
```

# Support Vector Machine (SVM)

```python
svm = SVC()
svm.fit(X_train_scaled, y_train)
y_pred_svm = svm.predict(X_test_scaled)
```

```python
evaluate_model("Support Vector Machine (SVM)", y_test, y_pred_svm)
```

### Finding Optimal Parameters 

```python
param_grid_svm = {'C': [0.01, 0.1, 1, 10, 100], 'kernel': ['linear', 'rbf']}
grid_svm = GridSearchCV(svm, param_grid_svm, cv=5)
grid_svm.fit(X_train_scaled, y_train)
y_pred_svm = grid_svm.predict(X_test_scaled)
```

```python
grid_svm.best_estimator_
```

```python
evaluate_model("Support Vector Machine (SVM)", y_test, y_pred_svm)
```

# Decision Tree Classifier

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
y_pred_decision_tree = decision_tree.predict(X_test)
```

```python
evaluate_model("Decision Tree Classifier", y_test, y_pred_decision_tree)
```

### Finding Optimal Parameters 

```python
param_grid_decision_tree = {'max_depth': [None, 5, 10, 15]}
grid_decision_tree = GridSearchCV(decision_tree, param_grid_decision_tree, cv=5)
grid_decision_tree.fit(X_train_scaled, y_train)
y_pred_decision_tree = grid_decision_tree.predict(X_test_scaled)
```

```python
grid_decision_tree.best_estimator_
```

```python
evaluate_model("Decision Tree Classifier", y_test, y_pred_decision_tree)
```

# Random Forest Classifier

```python
random_forest = RandomForestClassifier()
random_forest.fit(X_train, y_train)
y_pred_random_forest = random_forest.predict(X_test)
```

```python
evaluate_model("Random Forest Classifier", y_test, y_pred_random_forest)
```

### Finding Optimal Parameters 

```python
param_grid_random_forest = {'n_estimators': [50, 100, 150], 'max_depth': [None, 5, 10]}
grid_random_forest = GridSearchCV(random_forest, param_grid_random_forest, cv=5)
grid_random_forest.fit(X_train_scaled, y_train)
y_pred_random_forest = grid_random_forest.predict(X_test_scaled)
```

```python
grid_random_forest.best_estimator_
```

```python
evaluate_model("Random Forest Classifier", y_test, y_pred_random_forest)
```

```python
adaboost = AdaBoostClassifier()
adaboost.fit(X_train_scaled, y_train)
y_pred_adaboost = adaboost.predict(X_test_scaled)

evaluate_model("AdaBoost Classifier", y_test, y_pred_adaboost)
```

```python
grid_adaboost.best_estimator_
```

```python
evaluate_model("AdaBoost Classifier", y_test, y_pred_adaboost)
```
